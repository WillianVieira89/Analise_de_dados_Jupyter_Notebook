Análise de Dados - Gráfico de Dispersão

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregamento da base de dados
base = pd.read_csv('trees.csv')
base.head()
```

```python
# Gera um gráfico de dispersão considerando o volume e a dispersão
plt.scatter(x = base.Girth, y = base.Volume, color = 'blue', facecolors = 'blue')
plt.title('Árvores')
plt.xlabel('Circuferência')
plt.ylabel('Volume')
```

```python
# Gráfico de linha considerando o "volume" eo atributo "girth" - usando a função "plot"
```

```python
plt.plot(base.Girth,base.Volume)
plt.title('Árvores')
plt.xlabel('Circuferência')
plt.ylabel('Volume')
```

```python
# Gráfico de dispersão com "afastamento" dos dados (jitter)
# O jitter adiciona uma aleatoriedade aos dados

#fit_reg - Linha de tendência
sns.regplot(x = base.Girth, y = base.Volume, data = base, x_jitter = 0.3, fit_reg = True)
```
