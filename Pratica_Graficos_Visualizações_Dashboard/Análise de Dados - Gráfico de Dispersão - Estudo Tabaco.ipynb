Análise de Dados - Gráfico de Dispersão - Estudo Tabaco

```python
# Importações
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Carregando de arquivo CSV
base2 = pd.read_csv('esoph.csv')
base2
```

```python
# Gráfico entre atributos 'alcgp' e 'ncontrols'
sns.catplot(x='alcgp', y='ncontrols', data=base2, jitter=False)
```

```python
# Gráfico entre atributos 'alcgp' e 'ncontrols' com agrupamento
sns.catplot(x='alcgp', y='ncontrols', data= base2, col='tobgp')
```
