Análise de Dados - Divisão de Telas (subgráficos)

```python
# Importação das bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregamento da base de dados
base = pd.read_csv('trees.csv')
base.head()
```

```python
# Criando gráfico com Girth e Volume
plt.scatter(base.Girth, base.Volume)
```

```python
# Criando gráfico com Girth e Volume
plt.scatter(base.Girth, base.Height)
```

```python
# Criando gráfico com Heigth e  Volume
plt.scatter(base.Height, base.Volume)
```

```python
# Criando um Histograma com Volume
plt.hist(base.Volume)
```

```python
# Imprimindo todos os gráficos juntos na mesma imagem

# Criação de figura, no qual os gráficos serão posicionados
plt.figure(1)
plt.subplot(2,2,1)
plt.scatter(base.Girth, base.Volume)
plt.subplot(2,2,2)
plt.scatter(base.Girth, base.Height)
plt.subplot(2,2,3)
plt.scatter(base.Height, base.Volume)
plt.subplot(2,2,4)
plt.hist(base.Volume)
```
