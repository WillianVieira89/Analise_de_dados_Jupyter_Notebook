Análise de Dados - Gráfico de Dispersão com seaborn

```python
# Importação das bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Carregamento da base de dados
base = pd.read_csv('co2.csv')
base.head()
```

```python
# Gráfico de dispersão utilizando os atribuitos conc e uptake, agrupamento pelo Type
sns.scatterplot(data=base, x='conc', y='uptake', hue='Type')
```

```python
#Seleção de registros específicos da base de dados (Quebec e Mississipi)
q = base.loc[base['Type'] == 'Quebec']
m = base.loc[base['Type'] == 'Mississippi'] 
```

```python
# Subgráfico com os dados de Quebec e Mississipi
plt.figure()
plt.subplot(1,2,1)
sns.scatterplot(data=q, x='conc', y='uptake').set_title('Quebec')
plt.figure()
plt.subplot(1,2,2)
sns.scatterplot(data=m, x='conc', y='uptake').set_title('Mississipi')
plt.show()
```

```python
# Mostrando os dados refrigerados e não refrigerados
ch = base.loc[base['Treatment'] == 'chilled']
nc = base.loc[base['Treatment'] == 'nonchilled']
```

```python
# Gráficos somente com 'chilled' e 'nonchilled'
plt.figure()
plt.subplot(1,2,1)
sns.scatterplot(data=ch, x='conc', y='uptake').set_title('Chilled')
plt.figure()
plt.subplot(1,2,2)  
sns.scatterplot(data=nc, x='conc', y='uptake').set_title('Non chilled')
plt.show()
```
