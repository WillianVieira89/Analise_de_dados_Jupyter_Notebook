Análise de Dados - Pandas

```python
import pandas as pd
```

```python
# Carrega arquivo para dataframe Pandas
dados = pd.read_csv('Credit.csv')

# Formato do dataframe
dados.shape
```

```python
# Resumo estatístico de colunas númericas
dados.describe()
```

```python
# Exibe os primeiros registros 
dados.head()
```

```python
# Exibe os últimos registros, com parâmetros
dados.tail(2)
```

```python
# Filtar por nome da coluna
dados[['duration']]
```

```python
# Filtra por linhas de indice
dados.loc[0:100:10]
```

```python
# Filtrar por linhas escolhendo as linhas
dados.loc[[1,3]]
```

```python
# Filtros
dados.loc[dados['purpose'] == "radio/tv"]
```

```python
# Outra condição
dados.loc[dados['credit_amount'] > 18000]
```

```python
# Atribuimos resultado a variável, criando outro dataframe
credito2 = dados.loc[dados['credit_amount'] > 18000]
print(credito2)
```

```python
# Definimos só algumas colunas
credito3 = dados[['checking_status', 'duration']].loc[dados['credit_amount'] > 18000]
print(credito3)
```

```python
# Séries, uma única coluna

# Pode ser criada a partir de listas, array do numpy ou coluna de dataframe
s1 = pd.Series([2,5,3,34,54,23,1,16])
print(s1)
```

```python
# Series a partir de um array do numpy
import numpy as np
array1 = np.array([2,5,3,34,54,23,1,16])
s2 = pd.Series(array1)
print(s2)
```

```python
# Series a partir de um dataframe
s3 = dados['purpose']
print(s3)
type(s3)
```

```python
# Note a diferença, temos um dataframe
d4 = dados[['purpose']]
print(d4)
type(d4)
```

```python
# Renomear objetos
dados.rename(columns={"duration":"duração","purpose":"propósito"})
```

```python
# Porém a alteração não é persistida
dados.head(1)
```

```python
# Para persistir a alteração 
dados.rename(columns={"duration":"duração","purpose":"propósito"}, inplace=True)
dados.head(1)
```

```python
# Excluir coluna]
dados.drop('checking_status',axis=1,inplace=True)
print(dados)
```

```python
dados.head(1)
```

```python
# Verificar dados nulos
dados.isnull()
```

```python
# Verificar dados nulos
dados.isnull().sum()
```

```python
# Retirar colunas com NaN
dados.dropna()
```

```python
# Preencher dados faltantes
dados['duração'].fillna(0,inplace = True)
```
