Análise de Dados - Gráfico de Barras e Setores

```python
# Importação de biblioteca
import pandas as pd
```

```python
# Carregamento da base de dados
base = pd.read_csv('insect.csv')
base.head()
```

```python
# Agrupamento dos dados no atributo 'spary', contando e somando os registros
agrupado = base.groupby(['spray'])['count'].sum()
agrupado
```

```python
# Gráfico de barras
agrupado.plot.bar(color = 'green')
```

```python
# Colocando cores nas barras 
agrupado.plot.bar(color = ['gray', 'yellow', 'red', 'pink', 'orange', 'brown'])
```

```python
# Montando um gráfico de pizza
agrupado.plot.pie()
```

```python
# Inserindo legenda num gráfico de pizza
agrupado.plot.pie(legend =  True)
```
